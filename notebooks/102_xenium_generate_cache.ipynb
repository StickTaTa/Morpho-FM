```python
from __future__ import annotations

from pathlib import Path
import os
import json

import torch

from st_pipeline.superres.feature_cache import ensure_feature_cache, get_common_genes
from st_pipeline.data.gene_vocab import load_gene_vocab, map_genes_to_vocab


ROOT = Path(os.environ.get("MORPHO_VC_ROOT", r"e:\Morpho-VC")).resolve()
CKPT_DIR = ROOT / "checkpoints" / "st_superres" / "xenium_cellfm"
CKPT_DIR.mkdir(parents=True, exist_ok=True)

# Slide IDs (match 101 notebook)
slide_ids = ["XEN1"]

# Cache params
force_recompute = True
max_dim = 16000
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Build feature cache (HE thumbnails + embeddings)
ensure_feature_cache(
    slide_ids=slide_ids,
    root=ROOT,
    force=force_recompute,
    device=device,
    max_dim=max_dim,
    use_tissue_seg=True,
)

# Build gene list (intersection) and map to CellFM vocab
common_genes = get_common_genes(slide_ids, root=ROOT)
vocab = load_gene_vocab(ROOT / "assets" / "cellfm" / "expand_gene_info.csv")
kept_genes, gene_ids = map_genes_to_vocab(common_genes, vocab)

(CKPT_DIR / "common_genes.json").write_text(json.dumps(kept_genes, indent=2), encoding="utf-8")
print(f"Saved common_genes.json with {len(kept_genes)} genes -> {CKPT_DIR}")
```
